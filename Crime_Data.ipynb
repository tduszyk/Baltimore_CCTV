```python
raw_data = raw_data[(raw_data["CrimeDate"] >= '2018-01-01 00:00:00') &
                           (raw_data["CrimeDate"] <= '2018-12-31 00:00:00')]
#recent_year_data['Neighborhood'].value_counts()
```
