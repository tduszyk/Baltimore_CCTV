```python
# import dependencies

import requests
import json
import pandas as pd
```

```python
# build URL for GET requests to retrieve data

# census base URL
base_url = "https://api.census.gov/data/"
# request 2017 American Community Survey 1-year data (most recent)
dataset = "2017/acs/acs5"
# specific data variables requested (total pop, white pop, black pop, median income)
variables = "?get=NAME,B01003_001E,B02008_001E,B02009_001E,B06011_001E"
# specfic location: by zip code
geography = "&for=zip%20code%20tabulation%20area:*"

url = base_url + dataset + variables + geography
print(requests.get(url))
```

```python
# retrieve data and convert it into JSON

#B01003_001E, B02008_001E,  B02009_001E, B06011_001E   
#total pop,   white pop,    black pop,   median income

print(requests.get(url).json())
# response_json = (requests.get(url).json())
# print(json.dumps(response_json, indent=4, sort_keys=True))
```
